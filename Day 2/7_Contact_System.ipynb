## Section 1: Define the Contact Class 

Steps to implement: 
1. Define a class named `Contact`.
 
2. Create an `__init__` method that takes `name`, `phone`, and `email` as parameters and assigns them to instance variables. 
3. Define a `__str__` method to return a formatted string representation of the contact. 

Display format: name | phone | email

## Section 2: Define the ContactBook Class 

Steps to implement: 
1. Import the `json` module.

2. Define a class named `ContactBook`.
3. Initialize an empty dictionary `contacts` in the constructor.
4. Implement the `add`method to add a contact to the dictionary only if the contact's name does not already exist as a key. 
    * Convert the contact's name to lowercase to use as the dictionary key.

    * Check if this lowercase name already exists in the contacts dictionary to prevent duplicates.

    * If the name is unique, add the contact object to the dictionary with the lowercase name as the key.

    * Print a success message confirming the contact was added.
    
    * If the name already exists, print an error message indicating a duplicate.

5. Implement the `search` method to find contacts by name using list comprehension over dictionary values.
    * Convert the search term to lowercase to ensure case-insensitive matching.

    * Iterate over the dictionary items to access both contact names (keys) and contact objects (values).

    * Check if the lowercase search term is contained within each lowercase contact name key.

    * Collect and return a list of all `Contact` objects that match the search criteria

6. Implement the `delete` method to remove contacts by name key.
    * Convert the input name to lowercase to match the dictionary keys.

    * Check if the lowercase name exists as a key in the contacts dictionary.

    * If the contact exists, delete it from the dictionary.

    * Print a confirmation message indicating the contact was deleted.

    * If the contact does not exist, print a message indicating no contact was found with that name.

7. Implement the `show_all` method to print all contacts or a message if none exist.
8. Create a generator method `iter_contacts` to yield contacts one by one from dictionary values.
9. Implement `save_to_file` to save contacts as JSON.
10. Implement `load_from_file` to load contacts from a JSON file into the dictionary.

## Section 3: Menu Loop

Steps to implement:

4. For option 1, prompt for contact details, create a `Contact`, and add it.

6. For option 3, prompt for a search keyword, call `search`, and display results.
