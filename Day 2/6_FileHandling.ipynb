# Write to file

```python
with open("example.txt", "w") as file:
  file.write("This is first line\n")
  file.write("This is second line\n")
```

# Read a File

```python
with open("example.txt") as file:
  print(file.read())
```

```python
with open("example.txt") as file:
  for line in file:
    print(line.strip())
```

# Append to file

```python
with open("example.txt", "a") as file:
    file.write("Appending a line.\n")
```
