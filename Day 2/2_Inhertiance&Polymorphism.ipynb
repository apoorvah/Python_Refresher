# Inheritance

* Refine the code by refactoring Person, Customer and Employee class
* Show Polymorshism
* hasinstance, hasattr
* class attributes
* MRO (Method Resolution Order)

```python
class Person:
  bank_name = "ABC Bank"
  def __init__(self,name, age, email):
    self.name = name
    self.age = age
    self.email = email

  def celebrate_birthday(self):
    self.age += 1
    return f"Happy Birthday, {self.name}! You are now {self.age}."

  def update_email_address(self, new_email):
    self.email = new_email
    return f"{self.name}'s email updated to {self.email}."

  def is_adult(self):
        return self.age >= 18

class Customer(Person):
  def __init__(self, name, age, email):
      super().__init__(name, age, email)
      self.balance = 0
      self.history = []

  def deposit(self, amount):
      if amount > 0:
        self.balance += amount
        self.history.append(f"+{amount}")
        print(f"Deposited {amount:.2f}")
      else:
        print("Enter a positive amount!")

  def withdraw(self, amount):
    if amount > self.balance:
        print("Insufficient funds!")
    elif amount > 0:
        self.balance -= amount
        self.history.append(f"-{amount}")
        print(f"Withdrew {amount:.2f}")
    else:
        print("Enter a positive amount!")

  def show_balance(self):
    print(f"Balance: {self.balance:.2f}")

  def show_history(self):
    print("Transaction History:")
    for item in self.history:
        print(item)

  def role(self):   # Polymorphism
        return f"{self.name} is a bank customer."

  def __str__(self):
      return f"Customer: {self.name}, Age: {self.age}, Balance: {self.balance}"

class Employee(Person):
  def __init__(self, name, age, email, position):
        super().__init__(name, age, email)
        self.position = position

  def approve_loan(self, customer_name, amount):
        return f"{self.name} ({self.position}) approved a loan of {amount} for {customer_name}."

  def role(self):
      return f"{self.name} is a {self.position} at the bank."

  def __str__(self):
      return f"Employee: {self.name}, Age: {self.age}, Position: {self.position}"
```

```python

# Demo run
people = [
    Customer("Alice", 30, "alice@example.com"),
    Employee("Bob", 40, "bob@example.com", "Manager"),
    Person("Charlie", 25, "char@example.com")  # plain Person
]

for p in people:
    print(p)  # __str__ will be called automatically
    if hasattr(p, "role"):
        print(p.role())
    if isinstance(p, Customer):
        print(p.deposit(200))
        print(p.withdraw(100))
    if isinstance(p, Employee):
        print(p.approve_loan("Alice", 2000))
    print("---")
```
