# Multiple Errors

```python
def divide(a, b):
    try:
        result = a / b
        print(result)
    except (TypeError, ZeroDivisionError) as e:
      print(e)


divide(10, 2)  # Result: 10.0 / divide() finished
divide(10, "abc")  # Error: cannot divide by zero! / divide() finished
```

# Extra functionalities

* try
* except
* else
* finally

```python
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: cannot divide by zero!")
        return None
    except TypeError as e:
        print(f"Type error: {e}")
        return None
    else:
        print(f"Result: {result}")   # runs if NO exception
        return result
    finally:
        print("divide() finished")  # always runs


```

```python
divide(10, 2)
```

```python
divide(10, 0)
```

```python
divide(10, "abc")
```

# Throwing errors

```python
class Customer:
  def __init__(self, name):
      self.name = name
      self.balance = 0
      self.history = []

  def deposit(self, amount):
      if amount > 0:
        self.balance += amount
        self.history.append(f"+{amount}")
        print(f"Deposited {amount:.2f}")
      else:
        print("Enter a positive amount!")

  def withdraw(self, amount):
    if amount > self.balance:
        raise ValueError("Insufficient funds!")
    elif amount > 0:
        self.balance -= amount
        self.history.append(f"-{amount}")
        print(f"Withdrew {amount:.2f}")
    else:
        print("Enter a positive amount!")

  def show_balance(self):
    print(f"Balance: {self.balance:.2f}")

  def show_history(self):
    print("Transaction History:")
    for item in self.history:
        print(item)

  def __str__(self):
    return f"{self.name} has {self.balance} rupess in their account"


c = Customer("Apoorva")
# c.withdraw(100)
```

# *Optional*

Custom Error

```python
# Define a custom exception
class InsufficientFundsError(Exception):
    def __init__(self, amount, balance):
        self.message = (
            f"Cannot withdraw {amount}. "
            f"Available balance: {balance}"
        )
        super().__init__(self.message)

# Use it in your code
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientFundsError(amount, self.balance)
        self.balance -= amount

acc = BankAccount(100)
try:
    acc.withdraw(500)
except InsufficientFundsError as e:
    print(e)    # Cannot withdraw 500. Available balance: 100
```
