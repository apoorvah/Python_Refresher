# **Functional Programming**

Functional programming is when you write your logic in terms of simple functions, each making a small transformation. These functions are applied one after the other. We can also pass these function as variables

## Lambda

```python
square = lambda x: x ** 2
print(square(5))


add = lambda a, b: a + b
print(add(3, 4))
```

## Map and Filter
* lazy evaluation - gives iterators
* Written in C
* Good for large data

```python
numbers = [1, 2, 3, 4, 5, 6]

list(map(lambda x: x**2, numbers))
```

```python
list(filter(lambda x: x%2==0, numbers))
```

## List Comprehension

* Eager execution
* C optimized

```python
numbers = [1, 2, 3, 4, 5, 6]
[x**2 for x in numbers]
```

```python
[x for x in numbers if x%2==0]
```

```python
[x if x%2 ==0 else x**2 for x in numbers]
```

```python
{x : x**2 for x in numbers if x%2 == 0}
```
