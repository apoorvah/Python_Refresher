```python
class Person:
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email

    def celebrate_birthday(self):
        self.age += 1
        return f"Happy Birthday, {self.name}! You are now {self.age}."

    def update_email(self, new_email):
        self.email = new_email
        return f"{self.name}'s email updated to {self.email}."

    def is_adult(self):
        return self.age >= 18
```

# Creating objects

```python
p1 = Person("Alice", 17, "alice@example.com")
p2 = Person("Bob", 25, "bob@example.com")
```

# Interacting with objects

```python
print(p1.is_adult())             # False
print(p1.celebrate_birthday())   # Happy Birthday, Alice! You are now 18.
print(p1.is_adult())             # True
print(p2.update_email("bob@newmail.com"))  # Bob's email updated to bob@newmail.com
```

```python
class Customer:
  def __init__(self, name):
      self.name = name
      self.balance = 0
      self.history = []

  def deposit(self, amount):
      if amount > 0:
        self.balance += amount
        self.history.append(f"+{amount}")
        print(f"Deposited {amount:.2f}")
      else:
        print("Enter a positive amount!")

  def withdraw(self, amount):
    if amount > self.balance:
        print("Insufficient funds!")
    elif amount > 0:
        self.balance -= amount
        self.history.append(f"-{amount}")
        print(f"Withdrew {amount:.2f}")
    else:
        print("Enter a positive amount!")

  def show_balance(self):
    print(f"Balance: {self.balance:.2f}")

  def show_history(self):
    print("Transaction History:")
    for item in self.history:
        print(item)

  def __str__(self):
    return f"{self.name} has {self.balance} rupess in their account"
```

```python
c1 = Customer("Apoorva")
c2 = Customer("Neha")
```

```python
c1.deposit(10)
```

```python
c1.show_history()
```

```python
c2.withdraw(10)
```
