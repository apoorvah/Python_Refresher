# Contact Management System - Beginner Project

This notebook guides you through building a simple Contact Management System in Python with clear step-by-step instructions for each section.

## Section 1: Define the Contact Class 

Steps to implement: 
1. Define a class named `Contact`.
 
2. Create an `__init__` method that takes `name`, `phone`, and `email` as parameters and assigns them to instance variables. 
3. Define a `__str__` method to return a formatted string representation of the contact. 

Display format: name | phone | email

```python
class Contact:
    def __init__(self, name, phone, email):
        self.name = name
        self.phone = phone
        self.email = email

    def __str__(self):
        return f"{self.name} | {self.phone} | {self.email}"
```

## Section 2: Define the ContactBook Class 

Steps to implement: 
1. Import the `json` module.

2. Define a class named `ContactBook`.
3. Initialize an empty dictionary `contacts` in the constructor.
4. Implement the `add`method to add a contact to the dictionary only if the contact's name does not already exist as a key. 
    * Convert the contact's name to lowercase to use as the dictionary key.

    * Check if this lowercase name already exists in the contacts dictionary to prevent duplicates.

    * If the name is unique, add the contact object to the dictionary with the lowercase name as the key.

    * Print a success message confirming the contact was added.
    
    * If the name already exists, print an error message indicating a duplicate.

5. Implement the `search` method to find contacts by name using list comprehension over dictionary values.
    * Convert the search term to lowercase to ensure case-insensitive matching.

    * Iterate over the dictionary items to access both contact names (keys) and contact objects (values).

    * Check if the lowercase search term is contained within each lowercase contact name key.

    * Collect and return a list of all `Contact` objects that match the search criteria

6. Implement the `delete` method to remove contacts by name key.
    * Convert the input name to lowercase to match the dictionary keys.

    * Check if the lowercase name exists as a key in the contacts dictionary.

    * If the contact exists, delete it from the dictionary.

    * Print a confirmation message indicating the contact was deleted.

    * If the contact does not exist, print a message indicating no contact was found with that name.

7. Implement the `show_all` method to print all contacts or a message if none exist.
8. Create a generator method `iter_contacts` to yield contacts one by one from dictionary values.
9. Implement `save_to_file` to save contacts as JSON.
10. Implement `load_from_file` to load contacts from a JSON file into the dictionary.

```python

import json

class ContactBook:
    def __init__(self):
        self.contacts = {}

    def add(self, name, phone, email):
        contact = Contact(name, phone, email)
        key = contact.name.lower()
        if key in self.contacts:
            print("Error: Contact with this name already exists!")
        else:
            self.contacts[key] = contact
            print(f"Added {contact.name}")

    def search(self, name):
        keyword = name.lower()
        return [contact for contact_name, contact in self.contacts.items() if keyword in contact_name]

    def delete(self, name):
        key = name.lower()
        if key in self.contacts:
            del self.contacts[key]
            print(f"Deleted contact named {name}")
        else:
            print(f"No contact found with name {name}")

    def show_all(self):
        if not self.contacts:
            print("No contacts found.")
        else:
            for c in self.contacts.values():
                print(c)

    def iter_contacts(self):
        for c in self.contacts.values():
            yield c

    def save_to_file(self, filename="contacts.json"):
        try:
            with open(filename, "w") as f:
                json.dump([c.__dict__ for c in self.contacts.values()], f)
            print("Contacts saved successfully.")
        except Exception as e:
            print(f"Error saving contacts: {e}")

    def load_from_file(self, filename="contacts.json"):
        try:
            with open(filename, "r") as f:
                data = json.load(f)
                self.contacts = {c["name"].lower(): Contact(**c) for c in data}
            print("Contacts loaded successfully.")
        except FileNotFoundError:
            print("No saved contacts found.")
        except Exception as e:
            print(f"Error loading contacts: {e}")
```

## Section 3: Menu Loop

Steps to implement:
1. Instantiate a `ContactBook` object.

2. Load contacts from file at the start.
3. Create an infinite loop to display a menu and get user input.
4. For option 1, prompt for contact details, create a `Contact`, and add it.
5. For option 2, iterate over contacts using the generator and print each.
6. For option 3, prompt for a search keyword, call `search`, and display results.
7. For option 4, prompt for a name and delete the matching contact.
8. For option 5, save contacts to file.
9. For option 6, exit the loop and program.
10. Handle invalid choices with a message.

```python
book = ContactBook()
book.load_from_file()  # Load saved contacts at start

while True:
    print("Menu: 1) Add 2) View All 3) Search 4) Delete 5) Save 6) Quit")
    choice = input("Choose an option: ")

    if choice == "1":
        name = input("Enter name: ")
        phone = input("Enter phone: ")
        email = input("Enter email: ")
        
        book.add(name, phone, email)
    elif choice == "2":
        # Use generator to iterate
        for contact in book.iter_contacts():
            print("Contact:", contact)
    elif choice == "3":
        keyword = input("Enter name to search: ")
        results = book.search(keyword)
        if results:
            for r in results:
                print(r)
        else:
            print("No matches found.")
    elif choice == "4":
        name = input("Enter name to delete: ")
        book.delete(name)
    elif choice == "5":
        book.save_to_file()
    elif choice == "6":
        print("Goodbye!")
        break
    else:
        print("Invalid choice, try again.")
```
