# Banking System Project
Goal: build a simple command‑line banking app with
* account creation
* deposits
* withdrawals
* balance checks
* transaction history.

Assumption: Each time the user interacts with our banking app, we create an account with zero balance

## Section 1: Welcome and Account creation
We’ll start by creating an account dictionary to store name, balance, and history.

1. Print a welcome message to greet the user.

2. Prompt the user to enter their name using the input() function.

3. Create an `account` dictionary with keys: `name, balance (initialized to 0.0), and history (an empty list)`.

4. Print a confirmation message showing the account has been created with the user's name.

## Section 2: Deposit Function
Add money to the account if the amount is positive.

Follow these steps to generate the deposit logic:

1. Define a function named `deposit` that takes an `amount` parameter.
2. Check if the `amount` is greater than zero.

3. If valid, add the `amount` to the account's balance.

4. Record the deposit in the account's transaction history as say +20.503

5. Print a confirmation message showing the deposited amount, upto two decimal places.

6. If the amount is not positive, print an error message.

## Section 3: Withdraw Function
Subtract money if balance is sufficient.

Follow these steps to generate the withdraw logic:

1. Define a function named `withdraw` that takes an `amount` parameter.

2. Check if the `amount` is greater than the current account balance.
3. If the `amount` exceeds the balance, print an "Insufficient funds!" message.
4. If the `amount` is positive and less than or equal to the balance, subtract it from the balance.
5. Record the withdrawal in the account's transaction history as say -15.00.
6. Print a confirmation message showing the withdrawn amount, up to two decimal places.
7. If the `amount` is not positive, print an error message.


## Section 4: Show Balance
Display the current balance.

Follow these steps to generate the show balance logic:

1. Define a function named `show_balance`.

2. Print the current balance from the account dictionary formatted to two decimal places.



## Section 5: Transaction History
Show all deposits and withdrawals.

Follow these steps to generate the transaction history logic:

1. Define a function named `show_history`.

2. Print a header "Transaction History:".
3. Loop through each item in the account's transaction history list.
4. Print each transaction item.

## Section 6: Menu Loop
Interactive menu using while + if/elif.

Follow these steps to generate the menu loop logic:

1. Create an infinite loop to continuously show the menu until the user chooses to quit.

2. Print the menu options: Deposit, Withdraw, Balance, History, and Quit.
3. Prompt the user to choose an option.
4. Use conditional statements to call the appropriate function based on the user's choice.
5. For Deposit and Withdraw, prompt the user to enter the amount and convert it to a float.
6. Call the corresponding function with the entered amount.
7. If the user chooses Quit, print a goodbye message and break the loop.
8. Handle invalid choices by printing an error message.
