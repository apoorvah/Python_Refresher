# Import

```python
import numpy as np
```

# # Create arrays

```python
a = np.array([10,12,31,9])
# a = np.array([10, 11.2, 2])
# a = np.array([10, 11.2, 2, "Hello"])
print(a)
print(type(a))
print(a.dtype)
print(type(a[0]))
print("Shape of a", a.shape, "This gives us the number of elements per dimension")
print("Num. of dimesnions", a.ndim)
print("Total number of elements in the array", a.size)
```

```python
b = np.array([[1, 2], [3, 4]])   # 2D
print(b)
print(type(b))
print(b.dtype)
print(type(b[0]))
print("Shape of b", b.shape, "This gives us the number of elements per dimension")
print("Num. of dimesnions", b.ndim)
print("Total number of elements in the array", b.size)
```

# Vectorization — no loops needed

```python
a = np.array([10,12,31,9])
b = np.ones_like(a)
print("a", a)
print("b", b)
print(a+b)
```

# Broadcasting

# Boolean Indexing

```python
a = np.array([10,12,31,9])
x = [True, False, True, False]
a[x]
```

```python
mask = a > 5
mask
```
