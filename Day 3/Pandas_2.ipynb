```python
import pandas as pd

# Load a CSV file
df = pd.read_csv("employee.csv")

# First look at the data
df.head()       # first 5 rows
df.tail(3)      # last 3 rows
df.shape        # (rows, cols) e.g. (1000, 8)

# Data types & missing values
df.info()       # column names, dtypes, nulls
df.describe()   # count, mean, std, min, max

# Select columns
df["salary"]             # one column (Series)
df[["name", "salary"]]   # multiple columns (DataFrame)

# Filter rows
df[df["salary"] > 50000]
df[df["dept"] == "Engineering"]


# Group and aggregate
dept_avg = df.groupby("dept")["salary"].mean()
print(dept_avg)

# Multiple aggregations
df.groupby("dept").agg({
    "salary": ["mean", "max"],
    "name":   "count"
})

# Value counts — frequency table
print(df["dept"].value_counts())

# Sorting
df.sort_values("salary", ascending=False).head(5)

# New column
df["salary_k"] = df["salary"] / 1000
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("employee.csv")

# Bar chart — salary by department
dept_avg = df.groupby("dept")["salary"].mean()
dept_avg.plot(kind="bar", color="steelblue")
plt.title("Avg Salary by Department")
plt.ylabel("Salary ($)")
plt.tight_layout()
plt.show()

# # Line chart — trend over time
# df.groupby("year")["salary"].mean().plot(kind="line")
# plt.title("Salary Trend")
# plt.show()

# Histogram — distribution
df["salary"].plot(kind="hist", bins=20, edgecolor="black")
plt.title("Salary Distribution")
plt.xlabel("Salary ($)")
plt.show()
```
